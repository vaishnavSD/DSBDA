```python
min_max_scaler=preprocessing.MinMaxScaler()
# min_max_scaler = preprocessing.MinMaxScaler()
x=df.iloc[:,:4]
x_scaled = min_max_scaler.fit_transform(x)
df_normalized = pd.DataFrame(x_scaled)
df_normalized
```

```python
df['species'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
df['species']= label_encoder.fit_transform(df['species'])
df['species'].unique()
df
```

```python
species = df[['species']]

# Initialize the OneHotEncoder
encoder = preprocessing.OneHotEncoder()

# Fit and transform the 'species' column
encoded_species = encoder.fit_transform(species).toarray()

# Create a DataFrame with the encoded species
encoded_df = pd.DataFrame(encoded_species, columns=encoder.categories_[0])

# Concatenate the encoded species DataFrame with the original DataFrame
df_encoded = pd.concat([df, encoded_df], axis=1)

# Display the encoded DataFrame
print(df_encoded.head())
encoded_species
encoded_df
df_encoded
```
